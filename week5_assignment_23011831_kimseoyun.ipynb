```python
# =========================
# Section 1 - Project Setup
# =========================

# 배너 출력 (문자 반복)
print("=" * 42)
print(" OpenTrack — Open Source Project")
print(" Project Lead : Kim seo yun")
print("=" * 42)

# 프로젝트 기본 정보 (변경 불가 → tuple 사용)
project = ("OpenTrack", "1.0", 2026, "Python", "Kim seo yun")

# 각 요소 인덱스로 출력
print(f"Name : {project[0]}")
print(f"Version : {project[1]}")
print(f"Started : {project[2]}")
print(f"Main Language : {project[3]}")
print(f"Project Lead : {project[4]}")

# 슬라이싱으로 앞 3개 출력
print(f"First 3 fields : {project[:3]}")

# tuple 메서드 사용
print(f"Language count : {project.count('Python')}")  # 특정 값 개수
print(f"Language index : {project.index('Python')}")  # 위치 찾기

# project[0] = "Test"
# → TypeError 발생 (tuple은 immutable, 값 변경 불가)
# → 프로젝트 기본 정보처럼 변하지 않는 데이터에 적합

# 기여자 리스트 초기화
contributors = []

# 기여자 4명 입력받기
for i in range(4):
    print(f"\nEnter contributor {i+1} information")

    # 사용자 입력
    name = input("Name: ")
    role = input("Role: ")
    language = input("Language: ")
    commits = int(input("Commits: "))
    country = input("Country: ")

    # dict 형태로 저장
    contributor = {
        "name": name,
        "role": role,
        "language": language,
        "commits": commits,
        "country": country
    }

    # 리스트에 추가
    contributors.append(contributor)

# 이름만 따로 리스트 추출
names = []
for contributor in contributors:
    names.append(contributor["name"])

# 정렬
names.sort()

print(f"\nSorted names : {names}")
print(f"Last name : {names[-1]}")   # 마지막 요소
print(f"First two : {names[:2]}")   # 앞 2개 슬라이싱

# 모든 contributor에 status 추가
for contributor in contributors:
    contributor.update({"status": "Active"})

# 첫 번째 contributor status 출력
print(f"Contributor 1 status : {contributors[0].get('status')}")

# 첫 contributor 복사 (백업)
backup = contributors[0].copy()
print(f"Backup : {backup}")


# =========================
# Section 2 - Issue Tracking
# =========================

# 이슈 리스트 초기화
issues = []

# 이슈 5개 입력
for i in range(5):
    print(f"\nEnter issue {i+1} information")

    issue_id = input("ID: ")
    title = input("Title: ")
    issue_type = input("Type (Bug/Feature): ")
    priority = input("Priority (Critical/High/Medium/Low): ")
    reporter = input("Reporter: ")
    status = input("Status (Open/In Progress/Resolved): ")

    # dict로 저장
    issue = {
        "id": issue_id,
        "title": title,
        "type": issue_type,
        "priority": priority,
        "reporter": reporter,
        "status": status
    }

    issues.append(issue)

# Open 상태 개수 직접 세기 (count() 사용 금지)
open_count = 0
for issue in issues:
    if issue["status"] == "Open":
        open_count += 1

print(f"\nOpen issues : {open_count}")

# 첫 번째 이슈 priority 변경
issues[0]["priority"] = "Critical"
print("First issue → priority updated to Critical.")

# 마지막 2개 슬라이싱 출력
print(f"Last two issues : {issues[-2:]}")

# -------------------------
# Set 연산
# -------------------------

# reporter 집합 생성
reporters = set()
for issue in issues:
    reporters.add(issue["reporter"])

# 기술 스택 집합 생성
tech_stack = set()
for contributor in contributors:
    tech_stack.add(contributor["language"])

# 요소 추가 및 제거
tech_stack.add("Go")        # 추가
tech_stack.discard("Ruby")  # 없으면 에러 없이 무시

# 집합 연산
union_set = reporters.union(tech_stack)
intersection_set = reporters.intersection(tech_stack)
difference_set = reporters.difference(tech_stack)

print("-" * 42)
print(f"reporters : {reporters}")
print(f"tech_stack : {tech_stack}")
print(f"union : {union_set}")
print(f"intersection : {intersection_set}")
print(f"difference : {difference_set}")

# priority 존재 여부 확인
all_priorities = set()
for issue in issues:
    all_priorities.add(issue["priority"])

if "Critical" in all_priorities:
    print("Critical present : YES — flag for immediate review.")
else:
    print("Critical present : NO")

# -------------------------
# Dictionary 연산
# -------------------------

# priority 개수 세기
priority_count = {}
for issue in issues:
    p = issue["priority"]
    if p in priority_count:
        priority_count[p] += 1
    else:
        priority_count[p] = 1

# 상태별 그룹핑
status_groups = {}
for issue in issues:
    s = issue["status"]
    t = issue["title"]
    if s in status_groups:
        status_groups[s].append(t)
    else:
        status_groups[s] = [t]

print("-" * 42)
print(f"priority keys : {priority_count.keys()}")
print(f"priority values : {priority_count.values()}")
print(f"priority items : {priority_count.items()}")

# 상태별 출력
for key, value in status_groups.items():
    print(f"Status Group → {key} : {value}")

# 최다 reporter 찾기 (max 사용 금지)
reporter_count = {}
for issue in issues:
    r = issue["reporter"]
    if r in reporter_count:
        reporter_count[r] += 1
    else:
        reporter_count[r] = 1

top_reporter = ""
top_count = 0

for reporter, count in reporter_count.items():
    if count > top_count:
        top_count = count
        top_reporter = reporter

print(f"Top reporter : {top_reporter} ({top_count} issues)")

# type 키 제거
removed_type = issues[0].pop("type")
print(f"Removed type : {removed_type}")
print(f"After pop('type') : {issues[0]}")


# =========================
# Section 3 - File Handling
# =========================

import os

# 폴더 이름 생성 (소문자 + 언더스코어)
folder_name = project[0].lower().replace(" ", "_")

# 폴더 존재 확인 후 생성
if not os.path.exists(folder_name):
    os.mkdir(folder_name)
    print(f"\nFolder created : {folder_name}/")
else:
    print(f"\nFolder already exists : {folder_name}/")

# 폴더 내부 파일 목록 출력
print(f"Folder contents : {os.listdir(folder_name)}")

# 파일 경로 생성
report_path = os.path.join(folder_name, "project_report.txt")
csv_path = os.path.join(folder_name, "issues.csv")

# 파일 쓰기 (예외 처리)
try:
    # txt 파일 작성
    with open(report_path, "w", encoding="utf-8") as file:
        file.write("=" * 42 + "\n")
        file.write(f"{project[0]} — PROJECT REPORT\n")

        # 프로젝트 정보 기록
        file.write(f"Project: {project}\n")
        file.write(f"Contributors: {contributors}\n")
        file.write(f"Issues: {issues}\n")

    # csv 파일 작성
    with open(csv_path, "w", encoding="utf-8") as file:
        file.write("id,title,priority,reporter,status\n")

        for issue in issues:
            file.write(f"{issue['id']},{issue['title']},{issue['priority']},{issue['reporter']},{issue['status']}\n")

    print(f"Files saved : {os.listdir(folder_name)}")

except IOError:
    print("파일 저장 오류 발생")

# 파일 읽기
try:
    # 전체 읽기
    print("\n--- read() ---")
    with open(report_path, "r", encoding="utf-8") as file:
        print(file.read())

    # 한 줄씩 읽기
    print("\n--- readline() ---")
    with open(report_path, "r", encoding="utf-8") as file:
        print(file.readline().strip())
        print(file.readline().strip())

    # 여러 줄 읽기
    print("\n--- readlines() ---")
    with open(report_path, "r", encoding="utf-8") as file:
        lines = file.readlines()

        print(f"Total lines : {len(lines)}")

        # 특정 키워드 포함 줄만 출력
        for line in lines:
            if "Critical" in line or "High" in line:
                print(line.strip())

except FileNotFoundError:
    print("파일 없음 오류 발생")


# =========================
# Bonus
# =========================

# Critical/High 이슈 제목만 추출
urgent = [i["title"] for i in issues if i["priority"] in ["Critical", "High"]]

print(f"\nUrgent issues : {urgent}")
print(f"Urgent count : {len(urgent)}")

try:
    # append 모드로 추가 작성
    with open(report_path, "a", encoding="utf-8") as file:
        file.write("\nURGENT ISSUES\n")
        for title in urgent:
            file.write(f"{title}\n")

    # 마지막 6줄 확인
    with open(report_path, "r", encoding="utf-8") as file:
        lines = file.readlines()
        print("\nLast 6 lines:")
        for line in lines[-6:]:
            print(line.strip())

except:
    print("append 오류")


# =========================
# Final Summary
# =========================

print("\n" + "=" * 42)
print(f"{project[0]} — FINAL SUMMARY")
print("=" * 42)

print(f"Project : {project[0]} Version : {project[1]} Lead : {project[4]}")
print(f"Contributors : {len(contributors)} Names : {names}")
print(f"Tech Stack : {tech_stack}")
print(f"Issues : {len(issues)} Open : {open_count}")
print(f"Top Reporter : {top_reporter} ({top_count})")

for k, v in priority_count.items():
    print(f"{k}:{v}")

print(f"Report : {report_path}")
print(f"CSV : {csv_path}")

print("=" * 42)
print("Project complete.")
print("=" * 42)
```
